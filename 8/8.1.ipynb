# Importing Libraries

```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

# Reading Dataset

```python
tweets = pd.read_table('onlyhash.data',header=None)
tweets.columns = ["username", "date", "memes"]
tweets.head(5)
```

# Calculating Entropy

```python
def entropy(count,n):
    entro = 0
    for i in count.values():
        f = i/n
        logf = np.log10(f)
        entro = entro + f * logf
    if(entro!=0):
        entro = entro*-1
    return entro
```

```python
dates = tweets['date'].unique()
grpbydate = tweets.groupby(tweets['date'])
```

# System Entropy

```python
sysent = []
for i in dates:
    eachgroup = grpbydate.get_group(i)
    N = eachgroup['username'].count()
    memes = Counter(eachgroup['memes'])
    sysent.append(entropy(memes,N))
print(sysent)
```

# Average User Entropy

```python
avg = []
for a in dates:
    ent = []
    eachgroup = grpbydate.get_group(a)
    usergroup = eachgroup.groupby(eachgroup['username'])
    users = eachgroup['username'].unique()
    for j in users:
        group = usergroup.get_group(j)
        n = group['username'].count()
        meme = Counter(group['memes'])
        ent.append(entropy(meme,n))
    avg.append(sum(ent)/len(ent))
print(avg)
```

# Plotting

```python
plt.xlabel("Rank")
plt.ylabel("Entropy")
plt.plot(sysent,'r.',avg, 'b.')
plt.show()
```

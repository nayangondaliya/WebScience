```python
#import required libraries
import pandas as pd
```

```python
#read dataset from csv and generate dataframe
df = pd.read_csv('data.csv')

#select required columns for histogram
df = df.loc[:, ['INTcount', 'EXTcount', 'URLfragments']]

#rename columns
df = df.rename(columns={"INTcount": "Internal", "EXTcount": "External", "URLfragments": "Local"})

#plot histogram
df.plot.hist(bins=range(0, 1800, 200), alpha=0.8, figsize = [15,8], edgecolor='black')
```

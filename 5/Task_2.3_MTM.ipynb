```python
#import required libraries
import pandas as pd
```

```python
#reading dataset from csv file and generating dataframe
df = pd.read_csv('data-1.csv')
```

```python
#generating new dataframe for timestamp column and removing null values from timestamp column
t_df = pd.DataFrame(df.loc[:, 'timestamp'].dropna(), columns = ['timestamp'])
```

```python
#getting dayhour 
t_df['dayhour'] = pd.to_datetime(t_df['timestamp'], errors='coerce').dt.hour

t_df.drop('timestamp', axis=1, inplace=True)

t_df = t_df.groupby('dayhour').size().reset_index(name='hour_freq')
```

```python
t_df.plot(x="dayhour", y="hour_freq", kind="bar", figsize=[15,8], width=0.4)
```
